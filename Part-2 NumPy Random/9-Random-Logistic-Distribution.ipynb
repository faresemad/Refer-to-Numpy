# Logistic Distribution

### يستخدم التوزيع اللوجستي لوصف النمو.
### تستخدم على نطاق واسع في التعلم الآلي في الانحدار اللوجستي والشبكات العصبية وما إلى ذلك.
### It has three parameters:

    loc - mean, where the peak is. Default 0.== (يعني ، حيث تكون الذروة. الافتراضي 0)

    scale - standard deviation, the flatness of distribution. Default 1 == (الانحراف المعياري ، تسطيح التوزيع. الافتراضي 1.).

    size - The shape of the returned array.

```python
#Draw 2x3 samples from a logistic distribution with mean at 1 and stddev 2.0
from numpy import random

x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)
```

## Visualization of Logistic Distribution -- تصور التوزيع اللوجيستى

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.logistic(size=1000), hist=False)

plt.show()
```

## Difference Between Logistic and Normal Distribution -- الفرق بين التوزيع اللوجستي والتوزيع العادي

### كلا التوزيعين متطابقان تقريبًا ، لكن التوزيع اللوجستي له مساحة أكبر تحت الذيل. بمعنى آخر. إنه يمثل احتمالًا أكبر لحدوث أحداث بعيدة عن المتوسط.

### للحصول على قيمة أعلى للمقياس (الانحراف المعياري) ، تكون التوزيعات العادية واللوجستية متطابقة تقريبًا بصرف النظر عن الذروة.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')

plt.show()
```
