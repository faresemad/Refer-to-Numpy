# Random Permutations

## Random Permutations of Elements == التبادل العشوائي للعناصر

## Shuffling Arrays

```python
#التبديل يعني تغيير ترتيب العناصر في مكانها. أي في المصفوفة نفسها.
from numpy import random
import numpy as np
arr = np.array([1, 2, 3, 4, 5]) #نعمل مصفوفه من العناصر التاليه
random.shuffle(arr) #هنا بنعمل تبادل عشوائي للعناصر
print(arr)
```

## Generating Permutation of Arrays   ===   توليد التقليب للصفوف

```python
from numpy import random
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
```
