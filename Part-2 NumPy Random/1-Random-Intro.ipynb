## Random Intro

```python
from numpy import random
```

```python
x = random.randint(100) #الناتج اى رقم ما بين ال100 و ال0
```

```python
print(x)
```

## Generate random float

```python
x = random.rand() #في هذه الحاله سيعطى رقم عشوائي نوعه float
print(x)
```

## Generate Random Array

```python
x = random.randint(100 , size=(5)) #مصفوفه فيها خمس عناصر قيمتهم تتراوح من 0 ال 100
print(x)
```

```python
x = random.rand(100) #مصفوفه فيها 100 رقم من النوع float
print(x)
```

```python
x = random.rand(3,5) # مصفوفه بها 3 مصفوفها كل مصفوفه تحتوى علي 5 ارقام من النوع float
print(x)
```

## Generate Random Number From Array

```python
x = random.choice([3, 5, 7, 9]) #توليد رقم عشوائي من المصفوفة
print(x)
```

```python
#ايضا ممكن استعمال الداله size
x = random.choice([3, 5, 7, 9], size=(3, 5)) #مصفوفه بها 3 مصفوفات كل مصفوفه تحتوى علي 5عناصر عشوائيه من المصفوفه 
print(x)
```
