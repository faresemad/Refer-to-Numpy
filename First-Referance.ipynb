# NumPy Data Type

```python
# string ------> "fares"
# integer ------> 1,2,3,...
# float ------> 1.5 , 2.3
# boolean -----> True , False
# complex ------> 1.0 + 2.0j , 1.5 + 2.5j
```

```python
# i -> integer
# b -> boolean
# c -> complex
# f -> float
# u -> unsighed integer
# M -> date time
# m -> time delta
# O -> Object
# S -> string
# U -> Unicode string
# V -> fixed shunk of memory for other type (void)
```
